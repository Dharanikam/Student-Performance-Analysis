# **Handle Missing Values**

 Numerical Columns:

Filling missing values with median

```python
num_cols = [
    "Age", "Attendance_Percentage", "Midterm_Marks",
    "Final_Exam_Marks", "Assignment_Score",
    "Internal_Assessment", "Study_Hours_per_Week", "Backlogs"
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col].fillna(df[col].median(), inplace=True)



df
```

```python
df['Backlogs'].value_counts(dropna=False)
```

# **Categorical Columns**

Filling missing values with mode:

```python
cat_cols = [
    "Gender", "Department", "Parent_Education_Level",
    "Internet_Access", "Extra_Curricular"
]

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

df
```

# **Fix Inconsistent Categorical Values**
Gender Column

```python
# Replace blank strings with NaN
df["Gender"] = df["Gender"].replace("", np.nan)


df["Gender"] = df["Gender"].replace({
    "M": "Male",
    "F": "Female",
    "male": "Male",
    "female": "Female"
})


df["Gender"].fillna(df["Gender"].mode()[0], inplace=True)

df
```

```python
df["Gender"].value_counts()
```

```python
# Remove extra spaces and standardize case
df["Internet_Access"] = df["Internet_Access"].str.strip().str.capitalize()
```

```python
df["Internet_Access"].value_counts()
```

```python
df["Extra_Curricular"] = df["Extra_Curricular"].str.strip().str.capitalize()
```

Check Invalid or Out-of-Range Values
Age

```python
df = df[(df["Age"] >= 17) & (df["Age"] <= 30)]
```

check datatypes

```python
df.dtypes
```

```python
df = df[(df["Age"] >= 17) & (df["Age"] <= 30)]
df = df.copy()
```

```python
df = df.assign(

    Backlogs=df["Backlogs"].astype(int),
    Age=df["Age"].astype(int),
    Study_Hours_per_Week=df["Study_Hours_per_Week"].astype(int)
)
```

```python

df.describe()
```

```python
df.duplicated().sum()
```

```python

df.dtypes
```

```python
#checking for null values again
df.isnull().sum()
```

To handle :
Year_of_Study	:95

```python
# Convert to numeric
df["Year_of_Study"] = pd.to_numeric(df["Year_of_Study"], errors="coerce")

# Fill missing values with most frequent year
df["Year_of_Study"].fillna(df["Year_of_Study"].mode()[0], inplace=True)
df.isnull().sum()
```

```python
num_cols = [
    "Age",
    "Attendance_Percentage",
    "Midterm_Marks",
    "Final_Exam_Marks",
    "Assignment_Score",
    "Internal_Assessment",
    "Study_Hours_per_Week",
    "Backlogs"
]
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower,
                np.where(df[col] > upper, upper, df[col]))
df[num_cols].skew()
```

**Skewness between -1 and +1 → Acceptable**

### Categorization of students based on their **performance**

```python
def performance_level(marks):
    if marks >= 75:
        return "Excellent"
    elif marks >= 60:
        return "Good"
    elif marks >= 40:
        return "Average"
    else:
        return "Poor"

df["Performance_Level"] = df["Final_Exam_Marks"].apply(performance_level)
df.head()
```

```python
df["Performance_Level"].value_counts()
```

```python
# Summary statistics
df.describe()
```

```python
# Department-wise average final marks
df.groupby("Department")["Final_Exam_Marks"].mean()
#eee students studied well
```

```python
#students who needs more improvement
at_risk_students = df[
    (df["Attendance_Percentage"] < 65) |
    (df["Final_Exam_Marks"] < 40)
]

at_risk_students.head()
```

```python
df["Recommendation"] = np.where(
    (df["Attendance_Percentage"] < 65) & (df["Study_Hours_per_Week"] < 10),
    "Needs Academic Support",
    "Normal"
)

df["Recommendation"].value_counts()
```
